# Tokenize

```python
from nltk import sent_tokenize
sentence_in_ExpTxt = sent_tokenize(example_text)
sentence_in_ExpTxt
```

```python
from nltk import word_tokenize

word_in_ExpTxt = word_tokenize(example_text)
word_in_ExpTxt
```

# Filter Words

```python
from nltk.corpus import stopwords 

stp = set(stopwords.words("english"))

fliterd = [a for a in word_in_ExpTxt if a.casefold() not in stp]
print(fliterd)
```

# Stemming

 Stemming is a technique used to reduce an inflected word down to its word stem. For example, the words “programming,” “programmer,” and “programs” can all be reduced down to the common word stem “program.” In other words, “program” can be used as a synonym for the prior three inflection words

```python
from nltk.stem.snowball import EnglishStemmer
from random import *
engStem = EnglishStemmer()
chword = choice(word_in_ExpTxt)
stmdword = engStem.stem(word=chword) 
print(f"Orginal => [ {chword} ], and  stemming => {stmdword}")
```

# Lemmatization 

Lemmatization is the process of grouping together different inflected forms of the same word. It's used in computational linguistics, natural language processing (NLP) and chatbots.

```python
from nltk.stem import WordNetLemmatizer
 
lmatizer = WordNetLemmatizer()

lmatid = {}
for a in word_in_ExpTxt:
    lmdt = lmatizer.lemmatize(a)
    if lmdt in lmatid:
        lmatid[lmdt] += [a]
    else:
        lmatid[lmdt] = [a]
print(lmatid)
```

Chunking

```python
import nltk
from nltk.chunk import RegexpParser
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')
# nltk.download('maxent_ne_chunker')
# nltk.download('words')
```

```python
# Define a chunk grammar (e.g., NP for noun phrase)
grammar = "NP: {<DT>?<JJ>*<NN>}"  # Determiner, optional adjectives, and a noun
chunk_parser = RegexpParser(grammar)

# Parse the POS tags
tree = chunk_parser.parse(postTAGed)
# tree.draw()  # Visualize the chunking (requires tkinter)
# print(tree)
```

chinking

Chinking removes or excludes certain words from a chunk. It's essentially the opposite of chunking.

```python
# Chunk grammar with chinking
grammar = """
NP: {<DT>?<JJ>*<NN>}      # Chunk rule for noun phrases
    }<VBZ|RB>{            # Chink rule to exclude verbs (VBZ) and adverbs (RB)
"""

chunk_parser = RegexpParser(grammar)

tree = chunk_parser.parse(postTAGed)

```
